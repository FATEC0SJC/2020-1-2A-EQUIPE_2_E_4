# Teste de completude - tabela STG_MVT_CRD

# 1. Importação das bibliotecas utilizadas

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

# 2. Criação do data frame para importação da tabela

```python
## Atribuindo o arquivo a uma variável(Dataframe)
mvt = pd.read_excel('STG_MVT_CRD.xlsx')
mvt
```

# 3. Verificando completude das colunas

```python
total = len(mvt)
cont1, cont2, cont3, cont4 = 0, 0, 0, 0
for i in range (len(mvt)):
    if (str(mvt.COD_MDL[i]) == 'E01' and str(mvt.VLR_SDO_UTZ_CRD_RTO) == ''):
        cont1 += 1
    if (str(mvt.COD_MDL[i]) == 'D01' and str(mvt.VLR_TOT_FAT) == ''):
        cont2 += 1
    if (str(mvt.COD_MDL[i]) == 'D01' and str(mvt.VLR_MIM_FAT) == ''):
        cont3 += 1
    if (str(mvt.VLR_PCL_FAT) == ''):
        if (str(mvt.COD_MDL[i]) != 'D01') and (str(mvt.COD_MDL[i]) != 'E01'):
            cont4 += 1

nulos = cont1 + cont2 + cont3 + cont4

total, nulos
```

# 4. Representação gráfica

```python
dados = ['Total', 'Nulos']
valores = [total, nulos]

barlist = plt.bar(dados,[total,nulos])
barlist[0].set_color('#FF8C00')
barlist[1].set_color('#FF0000')

plt.xticks(dados, fontsize=12)
plt.ylabel('Quantidade', fontsize=12)
plt.xlabel('Dados', fontsize=14)
plt.title('Completude dos dados', fontsize=18)

plt.show()
```
