## 1. Importação de bibliotecas

```python
import xlrd
import pandas as pd
import matplotlib.pyplot as plt
```

### 1. 2. Criação de um dataframe

```python
df1 = pd.read_excel('STG_MVT_CRD.xlsx')
```

```python
df2 = pd.read_excel('STG_MDL.xlsx')
```

# 2. Conformidade quanto ao códigos de modalidade válidos


### 2.1. Criação de uma lista dos códigos de modalidade

```python
lista=[]
for i in range (len(df2)):
    a= df2.COD_MDL[i]
    lista.append(a)
```

```python
lista
```

### 2.2. Verficação da tabela STG_MVT_CRD contém os códigos da modalidade.

```python
conf =len(df1.loc[df1['COD_MDL'].isin(lista)])
```

```python
# Verificação da quantidade de registros que possuem o código de modalidade válido.
conf
```

### 2.3. Verificação de quantidade de registro da tabela STG_MVT_CRD

```python
total=len(df1)
```

```python
total
```

## 2.4. Verificação de %  registros conforme baseado nos códigos de modalidade válidos


```python
n_conforme= (total-conf) *100/total
conforme = 100 - n_conforme
```

```python
n_conforme
```

```python
conforme
```

## 2.5. Representação gráfica

```python
labels = ['Conforme', 'Não Conforme']
sizes = [conforme, n_conforme]

colors = ['DeepSkyBlue','#00BFFF']
 
fig2, ax1 = plt.subplots(figsize = (10,8))
ax1.pie(sizes, colors = colors, labels=labels, autopct='%1.1f%%', startangle=30)

centre_circle = plt.Circle((0,0),0.60,fc='white')
fig = plt.gcf()
fig.gca().add_artist(centre_circle)

plt.title('Conformidade dos dados quanto ao código de modalidade válido', fontsize = 18)

ax1.axis('equal')  
plt.tight_layout()
plt.show()
```

## 3. Vericação se o campo DES_TIP_PSS está no padrão

### 3.1 Criação de lista para verificação de padrão

```python
## Criação de uma lista para verificar de existência de registros no padrão F(Pessoa Física)  e J(Pessoa Jurídica)
listafjM=['F','J']

## Criação de lista para verificação de existência de registros fora do padrão f e j ( letras minúsculas)
listafjm=['f','j']
```

```python
listafjM, listafjm
```

```python
## Atribuição a uma variável o números de registros válidos com F e J
conffjM =len(df1.loc[df1['DES_TIP_PSS'].isin(listafjM)])

## Atribuição a uma variável o números de registros fora do padrão f e j (letras minúsculas) 
conffjm =len(df1.loc[df1['DES_TIP_PSS'].isin(listafjm)])
```

```python
conffjM , conffjm
```

### 3.2 Cálculo de porcentagem de dados conforme e não conforme quanto à Pessoa Física ou Jurídica

```python
FJ = conffjM *100/total
fj = conffjm*100/total
difFJ = (total-conffjM)*100/total
```

### 3.3 Representação Gráfica

```python
labels = ['Total de Registros com F ou J', 'Total de Registros fora do padrão']
sizes = [FJ, difFJ]

colors = ['DeepSkyBlue','#00BFFF']
 
fig2, ax1 = plt.subplots(figsize = (10,8))
ax1.pie(sizes, colors = colors, labels=labels, autopct='%1.1f%%', startangle=30)

centre_circle = plt.Circle((0,0),0.60,fc='white')
fig = plt.gcf()
fig.gca().add_artist(centre_circle)

plt.title('Conformidade dos dados quanto a identificação de tipo de cliente padronizada', fontsize = 18)

ax1.axis('equal')  
plt.tight_layout()
plt.show()



labels = ['Total de Registros com F ou J', 'Total de Registros com f ou j']
sizes = [FJ, fj]

colors = ['DeepSkyBlue','#00BFFF']
 
fig2, ax1 = plt.subplots(figsize = (10,8))
ax1.pie(sizes, colors = colors, labels=labels, autopct='%1.1f%%', startangle=30)

centre_circle = plt.Circle((0,0),0.60,fc='white')
fig = plt.gcf()
fig.gca().add_artist(centre_circle)

plt.title('Conformidade dos dados quanto a identificação de tipo de cliente padronizada', fontsize = 18)

ax1.axis('equal')  
plt.tight_layout()
plt.show()
```

## 4. Verificação do campo ID_FNT_ITT da tabela STG_MVT_CRD se está de acordo com a tabela STG_FNT_ITT

### 4.1 Criação de um dataframe da tabela STG_FNT_ITT, para verificação de fontes válidas

```python
df3 = pd.read_excel('STG_FNT_ITT.xlsx')
```

### 4.2 Criação de uma lista das fontes existentes

```python
listafontes=[]
for i in range (len(df3)):
    a= df3.ID_STG_FNT_ITT[i]
    listafontes.append(a)
```

### 4.3 Verficação da tabela STG_MVT_CRD contém fontes válidas

```python
conf_fontes =len(df1.loc[df1['ID_FNT_ITT'].isin(listafontes)])
```

```python
conf_fontes
```

### 4.4 Verificação de % registros conforme baseado na identificação de fonte válidas

```python
fontes = conf_fontes *100/total
fontesnconf = (total-conf_fontes)*100/total
```

### 4.5 Representação gráfica

```python
labels = ['Fonte conforme', 'Fonte não Conforme']
sizes = [fontes, fontesnconf]

colors = ['DeepSkyBlue','#00BFFF']
 
fig2, ax1 = plt.subplots(figsize = (10,8))
ax1.pie(sizes, colors = colors, labels=labels, autopct='%1.1f%%', startangle=30)

centre_circle = plt.Circle((0,0),0.60,fc='white')
fig = plt.gcf()
fig.gca().add_artist(centre_circle)

plt.title('Conformidade dos dados quanto ao código da fonte válido', fontsize = 18)

ax1.axis('equal')  
plt.tight_layout()
plt.show()
```
