# Código para gerar a tabela única

## Importação das bibliotecas

```python
import pandas as pd
import csv
import numpy as np
import xlrd
```

## Criando dataframes para todas as tabelas

```python
df_end_pf = pd.read_excel('Endereço PF_NOVA.xlsx')
df_pf = pd.read_excel('PF_NOVA.xlsx')
df_mvt = pd.read_excel('MVT_NOVA.xlsx')
df_opr = pd.read_excel('OPR_NOVA.xlsx')
df_pgt = pd.read_excel('PGT_NOVA.xlsx')
```

## Criando a tabela que armazenará todos os dados para pesquisa

```python
CONSULTA = {'cpf':[],
            'id':[],
            'idc_sexo':[],
            'ano_dat_nascimento':[],
            'nom_cidade':[],
            'des_estado':[],
            'tip_cli':[],
            'id_opr_cad_pos':[],
            'qtd_pcl':[],
            'dat_vct_ult_pcl':[],
            'vlr_ctrd_fta_tfm':[],
            'sdo_ddr_tfm':[],
            'vlr_ctrd':[],
            'cod_mdl':[],
            'id_mvt_cad_pos':[],
            'id_pgt_cad_pos':[],
            'dat_pgt':[],
            'dat_vct_tfm':[],
            'vlr_pgt_tfm':[],
            'status':[]
           }
df_consulta = pd.DataFrame(CONSULTA, columns=['cpf','id','idc_sexo','ano_dat_nascimento','nom_cidade','des_estado',
                                              'tip_cli','id_opr_cad_pos','qtd_pcl','dat_vct_ult_pcl','vlr_ctrd_fta_tfm',
                                              'sdo_ddr_tfm','vlr_ctrd','cod_mdl','id_mvt_cad_pos','id_pgt_cad_pos',
                                              'dat_pgt','dat_vct_tfm','vlr_pgt_tfm', 'status'])
```

```python
df_consulta
```

## Colocando os dados na tabela única

```python
for i in df_pgt.index:
    id_opr = df_pgt.id_opr_cad_pos[i]
    df_consulta.loc[i, 'id_opr_cad_pos'] = df_pgt.loc[i, 'id_opr_cad_pos']
    df_consulta.loc[i, 'id_mvt_cad_pos'] = df_pgt.loc[i, 'id_mvt_cad_pos']
    df_consulta.loc[i, 'id_pgt_cad_pos'] = df_pgt.loc[i, 'id_pgt_cad_pos']
    df_consulta.loc[i, 'dat_pgt'] = str(df_pgt.loc[i, 'dat_pgt'])
    df_consulta.loc[i, 'dat_vct_tfm'] = df_pgt.loc[i, 'dat_vct_tfm']
    df_consulta.loc[i, 'vlr_pgt_tfm'] = df_pgt.loc[i, 'vlr_pgt_tfm']
        
    for j in df_opr.index:
        if df_opr.id_opr_cad_pos[j] == id_opr:
            cpf = df_opr.doc_cli[j]     
            df_consulta.loc[i, 'tip_cli'] = df_opr.loc[j, 'tip_cli']
            df_consulta.loc[i, 'qtd_pcl'] = df_opr.loc[j, 'qtd_pcl']
            df_consulta.loc[i, 'dat_vct_ult_pcl'] = df_opr.loc[j, 'dat_vct_ult_pcl']
            df_consulta.loc[i, 'vlr_ctrd_fta_tfm'] = df_opr.loc[j, 'vlr_ctrd_fta_tfm']
            df_consulta.loc[i, 'sdo_ddr_tfm'] = df_opr.loc[j, 'sdo_ddr_tfm']            
            df_consulta.loc[i, 'vlr_ctrd'] = df_opr.loc[j, 'vlr_ctrd']        
            df_consulta.loc[i, 'cod_mdl'] = df_opr.loc[j, 'cod_mdl']
            
            for k in df_pf.index:
                if df_pf.cpf[k] == cpf:
                    id_pf = df_pf.id[k]
                    df_consulta.loc[i, 'cpf'] = df_pf.loc[k, 'cpf']
                    df_consulta.loc[i, 'id'] = df_pf.loc[k, 'id']
                    df_consulta.loc[i, 'idc_sexo'] = df_pf.loc[k, 'idc_sexo']
                    df_consulta.loc[i, 'ano_dat_nascimento'] = df_pf.loc[k, 'ano_dat_nascimento']
                    
                    for l in df_end_pf.index:
                        if df_end_pf.id_pessoa_fisica[l] == id_pf:
                            df_consulta.loc[i, 'nom_cidade'] = df_end_pf.loc[l, 'nom_cidade']
                            df_consulta.loc[i, 'des_estado'] = df_end_pf.loc[l, 'des_estado'];
```

## Agrupando os dados por CPF

```python
df = df_consulta.sort_values(['cpf'])
```

```python
df.reset_index(level=0, inplace=True)
df_consulta2 = df.drop('index', axis=1)
```

```python
df_consulta2
```

## Organizando datas da coluna dat_vct_ult_pcl

```python
for i in df_consulta2.index:
    a = (str(df_consulta2.dat_vct_ult_pcl[i]))
    a = a[0:2] + '/' + a[2:4] + '/' + a[4:8] 
    df_consulta2.dat_vct_ult_pcl[i] = a
```

# Organizando datas da coluna dat_pgt

```python
for i in df_consulta2.index:
    a = (str(df_consulta2.dat_pgt[i]))
    if len(a)<8:
        a = '0'+(str(df_consulta2.dat_pgt[i])) 
    a = a[0:2] + '/' + a[2:4] + '/' + a[4:8] 
    df_consulta2.dat_pgt[i] = a
```

# Organizando datas da coluna dat_vct_tfm

```python
for i in df_consulta2.index:
    a = (str(df_consulta2.dat_vct_tfm[i]))
    b = a.replace('-', '').replace(':', '')
    b = b[0:-6]
    b = b[6:8] + "/" + b[4:6] + "/" + b[0:4]
    df_consulta2.dat_vct_tfm[i] = b
#dat_vct_list
```

# Comparando as colunas

```python
lista = []
for i in df_consulta2.index:
    if df_consulta2.dat_vct_tfm[i] == '//':
        df_consulta2.loc[i, 'status'] = 'APTO'
    else:
        dia_pgt = df_consulta2.dat_pgt[i][0:2]
        mes_pgt = df_consulta2.dat_pgt[i][3:5]
        ano_pgt = df_consulta2.dat_pgt[i][6:11]

        dia_vct = df_consulta2.dat_vct_tfm[i][0:2]
        mes_vct = df_consulta2.dat_vct_tfm[i][3:5]
        ano_vct = df_consulta2.dat_vct_tfm[i][6:11]

        if ano_pgt < ano_vct:
            df_consulta2.loc[i, 'status'] = 'APTO'
        else: 
            if ano_pgt == ano_vct:
                if mes_pgt < mes_vct:
                    df_consulta2.loc[i, 'status'] = 'APTO'
                else:
                    if mes_pgt == mes_vct:
                        if dia_pgt <= dia_vct:
                            df_consulta2.loc[i, 'status'] = 'APTO'
                        else: df_consulta2.loc[i, 'status'] = 'NÃO APTO'
                    else: df_consulta2.loc[i, 'status'] = 'NÃO APTO'
            else: df_consulta2.loc[i, 'status'] = 'NÃO APTO'
                
    if df_consulta2.status[i] == 'NÃO APTO':
        lista.append(df_consulta2.cpf[i])
```

```python
for i in df_consulta2.index:
    if df_consulta2.cpf[i] in lista:
        df_consulta2.loc[i, 'status'] = 'NÃO APTO'
```
