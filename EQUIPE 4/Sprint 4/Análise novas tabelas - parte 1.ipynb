## Importação das bibliotecas

```python
import pandas as pd
import csv
import numpy as np
import xlrd
```

## Função para transformar arquivos csv em dataframe

```python
def importacao_csv(arquivo):
  with open(arquivo,'r') as csv_file:
          csv_reader = csv.reader(csv_file)

          next(csv_reader)

          c = 0

          lista = []

          metadado_tratado = []

          for line in csv_reader:
              if(c == 0):
                  lista.append(line[0].split('|'))
              c += 1
        
          for item in lista[0]:
              if item:
                  metadado_tratado.append(str(item).replace(" ",""))

  with open(arquivo,'r') as csv_file:
      csv_reader = csv.reader(csv_file)

      next(csv_reader)

      c = 0

      lista = []

      corpo_tratado = []

      for line in csv_reader:
        if(c != 0):
          lista.append(line[0].split('|'))
        c +=1

      for item in lista:
        nova_lista = []
        for i in item:
          if(i):
            if '+--' not in i:
              nova_lista.append(str(i).replace(" ",""))
        corpo_tratado.append(nova_lista)
  data = corpo_tratado
  metadado = metadado_tratado

  data_tratado = []

  for i in data:
      cont = 0
      for z in i:
          data_tratado.append([])
          data_tratado[cont].append(z)
          cont += 1
  dicionario = dict(zip(metadado, data_tratado))
  a = pd.DataFrame.from_dict(dicionario)
  a = a.drop_duplicates()
  a = a.dropna()
        
  return a
```

```python
opr = 'fatec_opr.csv'
mvt = 'fatec_mvt.csv'
pgt = 'fatec_pgt.csv'
end_pf = 'fatec_endereco_pessoa_fisica.csv'
pf = 'fatec_pessoa_fisica.csv'
```

## Transformação dos arquivos csv para dataframe

```python
df_opr = importacao_csv(opr)
df_mvt = importacao_csv(mvt)
df_pgt = importacao_csv(pgt)
df_end_pf = importacao_csv(end_pf)
df_pf = importacao_csv(pf)
```
